```python
# Importing display from IPython library
from IPython.display import display

# Importing tokenizers
from nltk.tokenize import sent_tokenize, word_tokenize, PunktSentenceTokenizer
```

```python
from nltk.corpus import inaugural
```

```python
# Now let us create for our training and testing data using US Presidential Inaugural address by Geroge Washington
train_text =inaugural.raw('1789-Washington.txt')
test_text =inaugural.raw('1793-Washington.txt')
```

```python
costum_sent_tokenizer = PunktSentenceTokenizer(train_text)
tokenized_text = costum_sent_tokenizer.tokenize(test_text)
```

```python
# creating a function that will run through all the tokens and create Named Entities 
# corpus G. Washington Inaugural Address:
def process_content():
    try:
        for i in tokenized_text[:3]:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            namedEnt = nltk.ne_chunk(tagged, binary=True)   # setting binary=True
            display(namedEnt)

    except Exception as e:
        print(str(e))


# calling the fucntion
process_content()
```
