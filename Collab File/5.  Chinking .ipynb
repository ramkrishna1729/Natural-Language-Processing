```python
# Importing display from IPython library
from IPython.display import display

# Importing tokenizers
from nltk.tokenize import sent_tokenize, word_tokenize, PunktSentenceTokenizer
```

```python
# imporitng corpus - Inaugural
from nltk.corpus import inaugural
```

```python
# Now let us create for our training and testing data using US Presidential Inaugural address by Geroge Washington
train_text1 = inaugural.raw('1789-Washington.txt')
test_text1 = inaugural.raw('1793-Washington.txt')
```

```python
# We are training the Punkt tokenizer again with new training text like:
custom_sent_tokenizer1 = PunktSentenceTokenizer(train_text1)

# Using the testing text to actually tokenize the corpus "1793-Washington"
tokenized_text1 = custom_sent_tokenizer1.tokenize(test_text1)
```

```python
# creating a function that will run through all the tokens and create chunks 
# corpus G. Washington Inaugural Address:
def process_content():
    try:
        for i in tokenized_text1[:3]:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)

          # in this step we will try to chink the unwanted part-of-speech tags of tokens 
            chunkGram = r"""Chunk: {<.*>+}
                                    }<VB.?|IN|DT|TO>+{"""

            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            display(chunked)

    except Exception as e:
        print(str(e))


# calling the fucntion
process_content()
```
