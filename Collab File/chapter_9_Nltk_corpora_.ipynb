```python
import nltk
# Downloading nltk files
nltk.download_shell()
```

```python
#importing gutenberg Corpus
from nltk.corpus import gutenberg
```

```python
#taking look at the identifiers if this corpus
gutenberg.fileids()
```

```python
blake = gutenberg.raw('blake-poems.txt')
print(blake)
```

```python
len(gutenberg.words('blake-poems.txt'))
```

```python
from nltk.tokenize import sent_tokenize ,word_tokenize
```

```python
# Getting the list of sentences, where each sentence is itself is a list of words
gutenberg.sents('blake-poems.txt')
```

```python
tok = sent_tokenize(blake)
print(tok[5:15])
```

```python
blake_sent_tokens = sent_tokenize(blake)  
#now printing them in the range of 10
for sent_token in blake_sent_tokens[:10]:
  print(word_tokenize(sent_token)) 
```
